# formatting season name in player performances

```python
import re
import pandas as pd

# Load the file
df = pd.read_csv("player_performances_cleaned.csv")

def standardize_season(season):
    if pd.isna(season):
        return season

    season = str(season).strip()

    # Case 1: Already in YY/YY format (e.g., 08/09, 22/23)
    if re.match(r"^\d{2}/\d{2}$", season):
        return season

    # Case 2: YYYY format (e.g., 2020, 2019)
    if re.match(r"^\d{4}$", season):
        start = int(season) % 100
        end = (start + 1) % 100
        return f"{start:02d}/{end:02d}"

    # Case 3: YYYY/YYYY (rare but possible)
    if re.match(r"^\d{4}/\d{4}$", season):
        start = int(season[:4]) % 100
        end = (start + 1) % 100
        return f"{start:02d}/{end:02d}"

    return season

df['season_name'] = df['season_name'].apply(standardize_season)

# save cleaned file
df.to_csv("player_performances_cleaned.csv",index=False)
```

# Dropping the columns with higher missing values and formatting names.

```python
import pandas as pd
import re
import unicodedata

#load CSV
df= pd.read_csv("player_profiles.csv", low_memory=False)

# drop unwanted columns
cols_to_drop = ["date_of_death" ,"fourth_club_name", "fourth_club_url" ,"third_club_name" ,"third_club_url" ,"second_club_name" ,"second_club_url" ,
"contract_there_expires" ,"on_loan_from_club_name" ,
"on_loan_from_club_id" ,"date_of_last_contract_extension" ,"contract_option" ,"social_media_url", "outfitter" ,"player_slug" ,
"player_image_url" ,"name_in_home_country" ,"place_of_birth" ,"current_club_name" ,"player_agent_name" ,"player_agent_id" ,"current_club_id"]

df = df.drop(columns=cols_to_drop)

# Function to clean player_name
def clean_player_name(name):
    if pd.isna(name):
        return name
    name = str(name)

    name = re.sub(r"\(.*?\)", "", name)   # remove anything inside brackets: (1), (123), etc.
    name = re.sub(r"\d+", "", name)       # remove any numbers
    name = re.sub(r"^\*", "", name)       # remove * only if it is at the START
    name = unicodedata.normalize("NFKD", name)  # Remove accents / diacritics (  S,   U,   O)
    name = name.encode("ascii", "ignore").decode("utf-8")
    name = name.strip()                   # remove extra spaces
    name = name.upper()                   # convert to UPPERCASE
    
    return name

# Apply to player_name column
df["player_name"] = df["player_name"].apply(clean_player_name)

# save cleaned file
df.to_csv("player_profile_cleaned.csv",index=False)
```

# Merging the files:

```python
import pandas as pd

# 1. Load profile (small enough)
profile = pd.read_csv("player_profile_cleaned.csv", dtype={
    "player_id": "int32",
    "player_name": "category",
    "country_of_birth": "category",
    "citizenship": "category",
    "is_eu": "category",
    "position": "category",
    "main_position": "category",
    "foot": "category",
    "height": "float32"
})

# Convert date columns to datetime
date_cols = ["date_of_birth", "joined", "contract_expires"]
for col in date_cols:
    profile[col] = pd.to_datetime(profile[col], errors="coerce")

# 2. Load market value (keep latest value only)
mv = pd.read_csv("player_market_value.csv")

# convert types
mv["player_id"] = mv["player_id"].astype("int32")

# Detect date type and convert

# Case 1: column contains normal date strings like "2023-12-19"
if mv["date_unix"].dtype == object:
    # Try converting to datetime
    mv["date_unix"] = pd.to_datetime(mv["date_unix"], errors="coerce")
    # Convert to Unix timestamp (seconds)
    mv["date_unix"] = mv["date_unix"].astype("int64") // 10**9

# Case 2: already numeric (unix)
else:
    mv["date_unix"] = mv["date_unix"].astype("int64")

# Convert value column safely
mv["value"] = pd.to_numeric(mv["value"], errors="coerce").astype("float32")

# keep only latest market value by date_unix
mv_latest = mv.sort_values("date_unix").groupby("player_id").tail(1)

# 3. Load injuries
inj = pd.read_csv("player_injuries_cleaned.csv")
inj["player_id"] = inj["player_id"].astype("int32")

# remove extra duplicates (keep 10 max per player-season)
inj = inj.groupby(["player_id", "season_name"]).head(10)

# 5. Chunk merge for the 150 MB performance file
chunksize = 300000   

merged_chunks = []

for chunk in pd.read_csv("player_performances_cleaned.csv", chunksize=chunksize):
    
    # convert types
    chunk["player_id"] = chunk["player_id"].astype("int32")
    chunk["season_name"] = chunk["season_name"].astype("category")
    chunk["competition_name"] = chunk["competition_name"].astype("category")
    chunk["team_name"] = chunk["team_name"].astype("category")
    
    #  MERGE Step 1: profile
    temp = chunk.merge(profile, on="player_id", how="left")
    
    #  MERGE Step 2: market value
    temp = temp.merge(mv_latest, on="player_id", how="left")
    
    #  MERGE Step 3: injuries
    temp = temp.merge(inj, on=["player_id","season_name"], how="left")
    
    merged_chunks.append(temp)

# concatenate all processed chunks
final_df = pd.concat(merged_chunks, ignore_index=True)

# Save final merged file
final_df.to_csv("merged_without_sentiment.csv", index=False)
```

# Merging the merged_output with the sentiment file:

```python
import pandas as pd

# Paths
big_path = "merged_without_sentiment_drop_name.csv"      
sent_path = "sentiment_cleaned_name.csv"      
output_path = "final_merged_drop_name.csv" 

# Load sentiment file
sent = pd.read_csv(sent_path, parse_dates=["tweet_date", "game_date"])

# Keep only necessary columns to save memory
sent = sent[[
    "player_name", "text", "vader_polarity", "vader_emotion",
    "tb_polarity", "tb_emotion", "game_date", "tweet_date", "when"
]]

# Optional: keep only the latest sentiment per player per game_date to reduce duplicates
sent = sent.sort_values("tweet_date").groupby(["player_name", "game_date"]).tail(1).reset_index(drop=True)

# Chunk-wise merge
chunksize = 50000  
first = True

for chunk in pd.read_csv(big_path, chunksize=chunksize, parse_dates=["joined","contract_expires","from_date","end_date"]):
    
    # Merge with sentiment using raw player_name
    merged_chunk = chunk.merge(
        sent,
        on="player_name",
        how="left",
        suffixes=("", "_sent")
    )
    
    # Write merged chunk to CSV (append after first chunk)
    merged_chunk.to_csv(output_path, mode="w" if first else "a", header=first, index=False)
    first = False

    print("Processed chunk:", len(chunk))

print("All chunks processed. Final merged file saved to:", output_path)
```

# understanding the features by graphical representations.

```python
# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load the dataset
file_path = "final_merged_drop_name.csv"
df = pd.read_csv(file_path)

# 3. Define columns for pie charts
columns = [
    'is_eu',
    'foot',
    'main_position',
    'vader_emotion',
    'tb_emotion',
    'when',
    'own_goals',
    'second_yellow_cards',
    'direct_red_cards'
]

# 4. Filter columns with  9 unique values
pie_columns = [col for col in columns if col in df.columns and df[col].nunique(dropna=True) <= 9]

print("Columns plotted:", pie_columns)

# 5. Create subplot layout
n_cols = 3
n_rows = (len(pie_columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 6 * n_rows))
axes = axes.flatten()

# 6. Plot pie charts
for i, col in enumerate(pie_columns):
    counts = df[col].dropna().value_counts()
    explode = [0.05] * len(counts)

    counts.plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        explode=explode,
        ax=axes[i]
    )

    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_ylabel("")

# 7. Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# 8. Save the figure
save_path = "pie_plots.png"  
fig.savefig(save_path, dpi=300)  
```

```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec

# 2. Load dataset
file_path = "final_merged_drop_name.csv"
df = pd.read_csv(file_path)

# 3. Define columns
columns = [
    'nb_in_group','nb_on_pitch', 'goals', 'assists',
    'subed_in', 'subed_out', 'yellow_cards',
    'penalty_goals', 'clean_sheets', 'position'
]

# 4. Filter valid columns
bar_columns = [
    col for col in columns
    if col in df.columns and df[col].dropna().shape[0] > 0
]

# 5. Grid setup

n_cols = 3
n_rows = (len(bar_columns) - 1 + n_cols - 1) // n_cols  # exclude last column

fig = plt.figure(figsize=(18, 6 * n_rows + 3))
gs = gridspec.GridSpec(n_rows + 1, n_cols, figure=fig)

# 6. Plot all columns except last (normal width)
for i, col in enumerate(bar_columns[:-1]):
    row = i // n_cols
    col_idx = i % n_cols
    ax = fig.add_subplot(gs[row, col_idx])

    summary = df[col].dropna().value_counts().sort_index()
    ax.bar(summary.index, summary.values)
    ax.set_title(f"Bar Plot of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")

# 7. Plot LAST column (full width)
last_col = bar_columns[-1]
ax_last = fig.add_subplot(gs[n_rows, :])  

summary = df[last_col].dropna().value_counts()
ax_last.bar(summary.index, summary.values)
ax_last.set_title(f"Bar Plot of {last_col}")
ax_last.set_xlabel(last_col)
ax_last.set_ylabel("Count")

# rotate labels so they dont overlap
ax_last.tick_params(axis='x', rotation=45)

# 8. Layout, show & save
plt.tight_layout()
plt.show()

fig.savefig("bar_plots.png", dpi=300)
```
