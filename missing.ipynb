```python
import pandas as pd

# Input and output file paths
input_path = "final_merged_date.csv"
output_path = "final_missing_handled_file.csv"

# Load the file
df = pd.read_csv(input_path)


# Columns you want to drop
columns_to_drop = [
    "country_of_birth",
    "from_date",
    "end_date",
    "when"
]

# Drop the columns safely (ignores missing ones)
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Save the cleaned file
df.to_csv(output_path, index=False)

print("File cleaned and saved as:", output_path)
```

```python
# Load the file
df = pd.read_csv("final_missing_handled_file.csv")

date_cols = [
    'date_of_birth','joined','contract_expires','date_unix','tweet_date', 'game_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')


num_cols = [
    'player_id','nb_in_group','nb_on_pitch','goals','assists','own_goals',
    'subed_in','subed_out','yellow_cards','second_yellow_cards',
    'direct_red_cards','penalty_goals','minutes_played',
    'goals_conceded','clean_sheets','height','days_missed','games_missed','tb_polarity','vader_polarity','value'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')


text_cols = [
    'player_name','citizenship','position',
    'main_position','team_name','competition_name',
    'injury_reason','text','season_name'
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()


categorical_cols = ['foot', 'vader_emotion', 'tb_emotion']
for col in categorical_cols:
    df[col] = df[col].astype(str)
for col in categorical_cols:
    df[col] = df[col].str.strip()
for col in categorical_cols:
    df[col] = df[col].str.lower()


# Save the cleaned file
df.to_csv("final_missing_handled_file.csv", index=False)
```

```python
# Load your dataset
df = pd.read_csv("final_missing_handled_file.csv")

# List of categorical columns to fill
categorical_cols = ['foot', 'position', 'citizenship']  # add more if needed

for col in categorical_cols:
    # Convert to string and strip spaces
    df[col] = df[col].astype(str).str.strip().str.lower()
    
    # Compute mode (most frequent value)
    mode_value = df[col].mode()[0]
    
    # Fill missing or empty values with mode
    df[col] = df[col].replace(['nan', '', 'none', 'na'], pd.NA)  # standardize missing
    df[col] = df[col].fillna(mode_value)

# Optional: check unique values after filling
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values, mode={df[col].mode()[0]}")

# Save the cleaned dataset
df.to_csv("final_missing_handled_file.csv", index=False)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("final_missing_handled_file.csv")

# Numerical columns
numeric_cols = ['minutes_played']

for col in numeric_cols:
    # Ensure numeric (invalid values → NaN)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Compute mode
    mode_value = df[col].mode(dropna=True)
    
    # Fill missing values with MODE (if exists)
    if not mode_value.empty:
        df[col] = df[col].fillna(mode_value[0])

# Save cleaned dataset
df.to_csv("final_missing_handled_file.csv", index=False)
```

```python
#Load dataset
df = pd.read_csv("final_missing_handled_file.csv")
import numpy as np
import pandas as pd

# Numeric columns
numeric_cols = ['value']

for col in numeric_cols:
    # Ensure numeric
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # 1️⃣ Handle missing values (mean)
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
    
# Save cleaned dataset
df.to_csv("final_missing_handled_file.csv", index=False)
```
